```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.svm import SVC

import utils
```

```python
iris = datasets.load_iris()
x = iris['data'][:, (2, 3)]  # petal length, petal width
y = iris['target']
setosa_or_versicolor = (y == 0) | (y == 1)
x = x[setosa_or_versicolor]
y = y[setosa_or_versicolor]
print(f"x.shape = {x.shape}, y.shape = {y.shape}")
```

```python
x_outliers = np.array([[3.4, 1.3], [3.2, 0.8]])
y_outliers = np.array([0, 0])
xo1 = np.concatenate([x, x_outliers[:1]], axis=0)
yo1 = np.concatenate([y, y_outliers[:1]], axis=0)
xo2 = np.concatenate([x, x_outliers[1:]], axis=0)
yo2 = np.concatenate([y, y_outliers[1:]], axis=0)
```

```python
svm_clf = SVC(kernel='linear', C=10**9)
svm_clf.fit(xo1, yo1)
utils.plot_decision_boundary(svm_clf, xo1, yo1, [0, 5.5, 0, 2]) # Impossible
plt.scatter(xo1[-1, 0], xo1[-1, 1], c='r', marker='x')
plt.show()
```

```python
svm_clf = SVC(kernel='linear', C=10**9)
svm_clf.fit(xo2, yo2)
utils.plot_decision_boundary(svm_clf, xo2, yo2, [0, 5.5, 0, 2])
plt.scatter(xo2[-1, 0], xo2[-1, 1], c='r', marker='x')
plt.show()
```

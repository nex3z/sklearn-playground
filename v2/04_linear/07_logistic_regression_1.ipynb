```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn import datasets
from sklearn.linear_model import LogisticRegression
```

```python
SEED = 42
```

```python
iris = datasets.load_iris()
```

```python
x = iris['data'][:, 3:]  # petal width
y = (iris['target'] == 2).astype(np.int)  # 1 if Iris virginica, else 0
print(f"x.shape = {x.shape}, y.shape = {y.shape}")
```

```python
log_reg = LogisticRegression(solver='lbfgs', random_state=SEED)
log_reg.fit(x, y)
```

```python
y_pred = log_reg.predict([x[0]])
y_prob = log_reg.predict_proba([x[0]])
print(f"y = {y[0]}")
print(f"y_pred = {y_pred}")
print(f"y_prob = {y_prob}")
```

```python
x_new = np.linspace(0, 3, 1000).reshape(-1, 1)
y_new_prob = log_reg.predict_proba(x_new)
decision_boundary = x_new[y_new_prob[:, 1] >= 0.5][0]
print(f"decision_boundary = {decision_boundary}")

plt.plot(x_new, y_new_prob[:, 1], label="Iris virginica")
plt.plot(x_new, y_new_prob[:, 0], label="Not Iris virginica")
plt.xlabel("Petal width")
plt.legend()
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

np.random.seed(42)
```

```python
from sklearn.datasets import make_moons

x, y = make_moons(n_samples=100, noise=0.25, random_state=53)

def plot_data(x=x, y=y, axis=[-1.5, 2.5, -1, 1.5]):
    plt.plot(x[:, 0][y==0], x[:, 1][y==0], '+')
    plt.plot(x[:, 0][y==1], x[:, 1][y==1], 'x')
    plt.axis(axis)
    
plot_data()
```

```python
def plot_decision_boundary(clf, axis=[-1.5, 2.5, -1, 1.5]):
    x1s = np.linspace(axis[0], axis[1], 100)
    x2s = np.linspace(axis[2], axis[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    x_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(x_new).reshape(x1.shape)

    cs = plt.contourf(x1, x2, y_pred, alpha=0.3)
    cbar = plt.colorbar(ticks=[0, 1, 2])
```

```python
from sklearn.tree import DecisionTreeClassifier

deep_tree_clf1 = DecisionTreeClassifier(random_state=42)
deep_tree_clf1.fit(x, y)

plt.figure()
plot_data()
plot_decision_boundary(deep_tree_clf1)
plt.show()
```

```python
deep_tree_clf2 = DecisionTreeClassifier(min_samples_leaf=4, random_state=42)
deep_tree_clf2.fit(x, y)

plt.figure()
plot_data()
plot_decision_boundary(deep_tree_clf2)
plt.show()
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

np.random.seed(42)
```

```python
m = 200
x = np.random.rand(m, 1)
y = 4 * (x - 0.5) ** 2 + np.random.randn(m, 1) / 10

def plot_data(axis=[0, 1, -0.2, 1]):
    plt.plot(x, y, '.')
    plt.axis(axis)
    
plot_data()
```

```python
def plot_decision_boundary(clf, axis=[0, 1, -0.2, 1]):
    xs = np.linspace(axis[0], axis[1], 500).reshape(-1, 1)
    y_pred = clf.predict(xs).reshape(xs.shape)
    plt.plot(xs, y_pred, 'r')
    plt.axis(axis)
```

```python
from sklearn.tree import DecisionTreeRegressor

tree_reg1 = DecisionTreeRegressor(random_state=42, max_depth=2)
tree_reg1.fit(x, y)

plt.figure()
plot_data()
plot_decision_boundary(tree_reg1)
plt.show()
```

```python
tree_reg2 = DecisionTreeRegressor(random_state=42, max_depth=3)
tree_reg2.fit(x, y)

plt.figure()
plot_data()
plot_decision_boundary(tree_reg2)
plt.show()
```

```python
tree_reg1 = DecisionTreeRegressor(random_state=42)
tree_reg1.fit(x, y)

plt.figure()
plot_data()
plot_decision_boundary(tree_reg1)
plt.show()
```

```python
tree_reg2 = DecisionTreeRegressor(random_state=42, min_samples_leaf=10)
tree_reg2.fit(x, y)

plt.figure()
plot_data()
plot_decision_boundary(tree_reg2)
plt.show()
```
